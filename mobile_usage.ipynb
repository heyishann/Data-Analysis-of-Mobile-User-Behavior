```python
df1 = df.loc[:,['App Usage Time (min/day)','Screen On Time (hours/day)', 'Battery Drain (mAh/day)', 'Number of Apps Installed', 'Data Usage (MB/day)', 'Age', 'User Behavior Class']]
df1
```

```python
N, F = df1.shape
N, F
```

```python
grouped = df1.groupby('User Behavior Class')
plt.bar([grp_name for grp_name, grp_item in grouped],[grp['Age'].mean() for grp_name , grp in grouped],yerr=[grp['Age'].std() for grp_name , grp in grouped] , width = 0.6, color=['r','g','b','y','pink'])
plt.xlabel('User Bahavior Class', fontweight='bold')
plt.ylabel('Average Age', fontweight='bold')
plt.title('Average age of the user according to user behavior class')
plt.show()
```

```python
fig, ax = plt.subplots(2,3, figsize=(14,9))
ax = ax.ravel()
for i in range(F-1):
  ax[i].hist(df1.iloc[:,i], bins=20)
  ax[i].set_xlabel(df1.columns[i], fontweight='bold')
  ax[i].set_ylabel('Count', fontweight='bold')
plt.show()
```

```python
fig , ax = plt.subplots(5,5 , figsize=[17,13])

for i in range(F-2):
  ax[i][i].text(0.05,0.5,df1.columns[i], fontweight='bold')
  for j in range(F-2):
    if i!=j:
      ax[i][j].scatter(df1.iloc[:,i], df1.iloc[:,j],s=30, c=df1.iloc[:,-1])
plt.show()
```

```python
grouped = df.groupby('User Behavior Class')
plt.pie([grp['Age'].mean() for grp_name , grp in grouped],labels = [grp_name for grp_name, grp_item in grouped], colors=['r','g','b','y','pink'],autopct='%1.2f%%')
plt.title('Average Age for Each User Behavior Class (Pie Chart)',fontweight='bold')
plt.show()
```

```python
os_count = df['Operating System'].value_counts()
for indx,data in enumerate(os_count):
  plt.bar(os_count.index[indx], data)
  plt.xlabel('Oprating Systems used by users', fontweight = 'bold')
  plt.ylabel('Numer of users', fontweight='bold')
plt.show()
```

```python
grouped = df.groupby('Gender')
for grp_name , grp in grouped:
  plt.bar(grp_name, grp['Data Usage (MB/day)'].mean(), width=0.3)
  plt.xlabel('Gender', fontweight='bold')
  plt.ylabel('Average Data Usage (MB/day)', fontweight='bold')
  plt.title('Average Data Usage (MB/day) according to Gender')
plt.show()
```

```python
fig , ax = plt.subplots(2,3 , figsize=[17,9])
ax = ax.ravel()
f = [(x - 0.5) / N for x in range (1,N+1)]
for i in range(F-1):
  ax[i].scatter(f , df1.iloc[:,i].sort_values(), c=df1.iloc[:,-1])
  ax[i].set_xlabel('Fi Values',fontweight='bold')
  ax[i].set_ylabel(df1.columns[i],fontweight='bold')
plt.show()
```

```python
device_counts = df['Device Model'].value_counts()
plt.pie(device_counts ,labels = device_counts.index, colors=['r','g','b','y','pink'],autopct='%1.2f%%')
plt.title('Distribution of Users according to Device used (Pie Chart)',fontweight='bold')
plt.show()
```

```python
grouped = df.groupby('Device Model')
for grp_name, grp in grouped:
  plt.bar(grp_name, grp['App Usage Time (min/day)'].mean())
  plt.xticks(rotation=45, fontsize=7)
  plt.xlabel('Device Model', fontweight='bold')
  plt.ylabel('Average App Usage Time (min/day)', fontweight='bold')
  plt.title('Devices used vs. Avg App usage Time')
plt.show()
```
